```python
import pandas as pd

# Load your dataset (replace 'your_dataset.csv' with the actual file name)
data = pd.read_csv('/content/spotify_songs.csv')

# Remove some columns (replace 'columns_to_remove' with the actual column names)
columns_to_remove = ['track_album_release_date', 'playlist_name', 'playlist_id', 'playlist_subgenre', 'track_album_id']
data = data.drop(columns=columns_to_remove, errors='ignore')

# Drop rows with null values in the 'song_name' column
data = data.dropna(subset=['track_name'])

# Change the name of a column to 'genre' (replace 'old_column_name' with the actual column name)
data = data.rename(columns={'playlist_genre': 'genre'})

data['user_id'] = range(1, len(data) + 1)

# Display the first few rows of the preprocessed data
print(data.head())

# Save the preprocessed data to a new CSV file if needed
data.to_csv('preprocessed_dataset.csv', index=False)
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from surprise import Dataset, Reader, KNNBasic
from surprise.model_selection import train_test_split
from surprise import accuracy

# Load the dataset (replace 'your_dataset.csv' with the actual file name)
data = pd.read_csv('/content/preprocessed_dataset.csv')

# Collaborative Filtering
reader = Reader(rating_scale=(1, 100))
surprise_data = Dataset.load_from_df(data[['user_id', 'track_id', 'track_popularity']], reader)
trainset, testset = train_test_split(surprise_data, test_size=0.2)
collaborative_model = KNNBasic(sim_options={'name': 'cosine', 'user_based': False})
collaborative_model.fit(trainset)

# Evaluate the model
predictions = collaborative_model.test(testset)
rmse = accuracy.rmse(predictions)
print(f'Root Mean Squared Error (RMSE) for Collaborative Filtering: {rmse}')

# Ask the user for input
user_input_genre = input("Enter your preferred genre: ")
user_input_song_name = input("Enter the name of a song you like: ")

# Content-Based Filtering
genre_songs = data[data['genre'] == user_input_genre]
user_input_song = data[data['track_name'] == user_input_song_name]

if user_input_song.empty:
    print(f"The song '{user_input_song_name}' was not found in the dataset.")
    print(f"Here are top recommendations in the genre '{user_input_genre}':")

    # Content-Based Recommendations for the specified genre
    features = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'valence', 'tempo']
    genre_songs_features = genre_songs[features]
    similarity_matrix_content = cosine_similarity(genre_songs_features)
    user_input_genre_index = genre_songs.index[0]  # Assuming there's only one genre input song
    similarity_scores_content = similarity_matrix_content[user_input_genre_index]
    content_based_recommendations = pd.DataFrame({'Song_Index': genre_songs.index, 'Similarity_Score_Content': similarity_scores_content})
    content_based_recommendations = content_based_recommendations.sort_values(by='Similarity_Score_Content', ascending=False)

    # Display top N recommendations (replace N with the desired number)
    top_recommendations = content_based_recommendations.head(5)
    # Print recommended song names
    recommended_song_names = data.loc[top_recommendations['Song_Index'].values, 'track_name'].tolist()
    print("Recommended Songs:")
    print(recommended_song_names)

else:
    # If the user-input song is found, proceed with the collaborative filtering recommendation
    user_input_song_index = trainset.to_inner_iid(user_input_song['track_id'].values[0])

    # Predict for the new user
    collaborative_recommendations = []
    for song_id in genre_songs['track_id']:
        predicted_score = collaborative_model.predict(uid='user_input_user_id', iid=song_id).est
        collaborative_recommendations.append({'track_id': song_id, 'Predicted_Score': predicted_score})

    # Display top N collaborative filtering recommendations (replace N with the desired number)
    collaborative_recommendations = pd.DataFrame(collaborative_recommendations)
    collaborative_recommendations = collaborative_recommendations.sort_values(by='Predicted_Score', ascending=False)
    top_collaborative_recommendations = collaborative_recommendations.head(5)

    # Print recommended song names
    recommended_song_names_collaborative = data.loc[top_collaborative_recommendations['track_id'].values, 'track_name'].tolist()
    print("Collaborative Filtering Recommendations:")
    print(recommended_song_names_collaborative)
```
